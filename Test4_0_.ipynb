```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#load
tips= sns.load_dataset('tips')
iris = sns.load_dataset('iris')
penguins = sns.load_dataset('penguins')
```

```python
print(iris)
```

```python
print(tips)
```

```python
print(penguins)
```

```python
#histogram of total_bill
sns.histplot(tips['total_bill'], bins=20)
plt.title("histogram")
plt.xlabel("total_bill")
plt.ylabel("frequency")
plt.show()
```

```python
#Create a scatterplot of total_bill vs tip and add hue for the sex column

sns.scatterplot(data=tips , x='total_bill', y='tip', hue='sex')
plt.title('scatterplot')
plt.show()
```

```python
#boxplot
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('boxplot')
plt.show()
```

```python
#countplot

sns.countplot(data=tips, x='day')
plt.title('countplot')
plt.show()
```

```python
#barplot
sns.barplot(data=tips, x='day', y='tip', estimator=np.mean)
plt.title('barplot for average')
plt.show()
```

```python
#kde
sns.kdeplot(data=iris, x='petal_length',fill=True)
plt.title(kdeplot)
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.show()
```

```python
sns.violinplot(tips, x='smoker', y='tip')
plt.show()
```

```python
corr= penguins.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

```python
#calculate Q1 and Q3

Q1 = np.percentile(tips['total_bill'],25)
Q3 = np.percentile(tips['total_bill'],75)

print(Q1)
print(Q3)
```

```python
IQR = Q3 - Q1

print(IQR)
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound)
print(upper_bound)
```
